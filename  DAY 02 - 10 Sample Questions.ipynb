# CONDITIONAL STATEMENTS 

### 1.CHECK POSITIVE,NEGATIVE OR ZERO
Take a number as input.

Print whether it is Positive, Negative, or Zero.

Also print whether it is Even or Odd

(only if it is not zero)

```python
num=float(input("Enter the Value:"))
if num>0:
    print("Positive")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
elif num<0:
    print("Negative")
    if num%2==0:
        print("Even")
    else:
        print("Odd")
else:
    print("The Number is either Zero or invalid ")
```

# 2. Ticket Price Calculator

A cinema charges ticket prices based on age:

  Below 5   → Free
  
  5 to 12   → ₹100
  
  13 to 17  → ₹150
  
  18 to 60  → ₹200
  
  Above 60  → ₹100 (Senior discount)

Take age as input and print the ticket price.

```python
age = int(input("Enter the Age: ")) 
if age<5:
    print("Free")
elif age<=12:
    print("The Ticket is ₹100")
elif age<=17:
    print("The Ticket is ₹150")
elif age<=60:
    print("The Ticket is ₹200")
elif age>=60:
    print("The Ticket is ₹100 (Senior Discount)")
else:
    print("Invalid Age..!!")
```

# 4.Triangle Validity & Type
Take 3 sides of a triangle as input.
First check if it is a valid triangle.
  Rule: sum of any 2 sides must be > third side
If valid → check type:
  All sides equal    → Equilateral
  Any 2 sides equal  → Isosceles
  All different      → Scalene
If not valid → print invalid message

```python
a=float(input("Enter the First side of the Triangle: "))
b=float(input("Enter the Second side of the Triangle: "))
c=float(input("Enter the Third side of the Triangle: "))
if (a+b>c) and (b+c>a) and (c+a>b):
    print("Valid Triangle")
    if a==b==c:
        print("Type : Equalateral")
    elif a==b or b==c or c==a:
        print("Type:  Isolated Triangle")
    else:
        print("Type : Scalene Triangle")
else:
    print("Not valid ")
```

# 5.Income  tax 
Calculate income tax based on annual income:
  0 to 2,50,000          → 0% tax
  2,50,001 to 5,00,000   → 5% tax
  5,00,001 to 10,00,000  → 20% tax
  Above 10,00,000        → 30% tax

Take income as input.
Print tax rate applied and total tax amount.

```python
income = float(input("Enter annual income: ₹"))

if income <= 250000:
    tax_rate = 0
    tax      = 0
elif income <= 500000:
    tax_rate = 5
    tax      = (income - 250000) * 0.05
elif income <= 1000000:
    tax_rate = 20
    tax      = (income - 500000) * 0.20
else:
    tax_rate = 30
    tax      = (income - 1000000) * 0.30

print(f"\nAnnual Income : ₹{income:,.2f}")
print(f"Tax Rate      : {tax_rate}%")
print(f"Tax Amount    : ₹{tax:,.2f}")
print(f"Net Income    : ₹{income - tax:,.2f}")
```

# 6.Sample Calculator
Take two numbers and an operator as input.
Operator can be: +  -  *  /  //  %  **

Perform the correct operation and print the result.
Handle division by zero with a proper message.

Example:
  Enter num1    : 10
  Enter operator: /
  Enter num2    : 2
  Result        : 10 / 2 = 5.0

```python
a=int(input("Enter the number 1: "))
op=input("Enter Operator : ")
b=int(input("Enter the number 2: "))
if op=="+":
    result=a+b
    print(f"Result: {a}+{b}={result}")
elif op=="-":
    result=a-b
    print(f"Result: {a}-{b}={result}")
elif op=="*":
    result=a*b
    print(f"Result: {a}*{b}={result}")
elif op=="-":
    result=a+b
    print(f"Result: {a}+{b}={result}")
elif op=="/":
    if b==0:
        print("Cannot able to divide the numde with zero ")
    else:
        result=a/b
        print(f"Result: {a}/{b}={result}")
elif op=="%":
    if b==0:
        print("Cannot able to divide the numde with zero ")
    else:
        result=a%b
        print(f"Result: {a}%{b}={result}")
elif op=="**":
    result=a**b
    print(f"Result: {a}**{b}={result}")
else:
    print("Invalid Input")
```

# 7.  Day of Week Checker
Take a number (1-7) as input.
Print the day of the week:
  1 → Monday    (Weekday)
  2 → Tuesday   (Weekday)
  3 → Wednesday (Weekday)
  4 → Thursday  (Weekday)
  5 → Friday    (Weekday — but also "Weekend is near! 🎉")
  6 → Saturday  (Weekend)
  7 → Sunday    (Weekend)
  Other → Invalid input
  

```python
number = int(input("Enter the Number from (1-7) as input: "))
match number:
    case 1:
        print("Monday (Weekday)")
    case 2:
        print("Tuesday (Weekday)")
    case  3:
        print("Wednesday (Weekday)")
    case  4:
        print("Thursday (Weekday)")
    case  5:
        print("Friday Weekday — but also Weekend is near")
    case  6:
        print("Saturday (Weekend)")
    case  7:
        print("Sunday (Weekend)")
    case _:
        print("Inavlid Input")
        
```
