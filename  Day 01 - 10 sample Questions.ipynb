# 3.SIMPLE CALCULATOR

Take two numbers as input from the user.

Perform and print ALL arithmetic operations:

  Addition       : 30
  
  Subtraction    : 10
  
  Multiplication : 200
  
  Division       : 2.0
  
  Floor Division : 2
  
  Modulus        : 0
  
  Power          : 100000

Note: Handle the case where second number is 0 (division by zero should print a warning)

```python
a=float(input("Enter the value of A:"))
b=float(input("Enter the value of B:"))
print(f"Addition : {a+b}")
print(f"Subtraction : {a-b}")
print(f"Multiplication : {a*b}")
if b!=0:
    print(f"Division: {a/b}")
    print(f"Floor Division: {a//b}")
    print(f"Modulus: {a%b}")
else:
    print("Cannot Divide by zero ")
print(f"Power: {a**b}")
```

# Q4 — Data Type Detective
You are given these 5 values:

  a = "100"
  
  b = 3.14

  c = True
  
  d = 0
  
  e = "45.67"

Tasks:

  1. Print the data type of each using type()
  2. Convert:
       a → int
     
       b → int  (what value do you get? why?)
     
       c → int
     
       d → bool
     
       e → float
     
  3. Print all converted values with their new types


```python
a = "100"
b = 3.14
c = True
d = 0
e = "45.67"
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
a=int(a)
b=int(b)
c=int(c)
d=bool(d)
e=float(e)
print(a,type(a))
print(b,type(b))
print(c,type(c))
print(d,type(d))
print(e,type(e))
```

# Q5 — Area & Perimeter Calculator
Take length and breadth of a rectangle as input.

Calculate and print:

  → Area      = length × breadth
  
  → Perimeter = 2 × (length + breadth)
  
  → Diagonal  = √(length² + breadth²)

Hint: For square root use → ** 0.5

Expected Output:

  Length     : 5.0
  
  Breadth    : 3.0
  
  Area       : 15.0 sq units
  
  Perimeter  : 16.0 units
  
  Diagonal   : 5.83 units

```python
length=float(input("Enter the Length: "))
breadth=float(input("Enter the Breadth: "))
area=length*breadth
perimeter=2*(length+breadth)
diagonal=(length**2+breadth**2)**0.5
print(f"Length: {length} Units")
print(f"Breadth: {breadth} Units")
print(f"Area: {area}sq units")
print(f"Perimeter: {perimeter}Units")
print(f"Diagonal: {diagonal:.2f}Units")
```
