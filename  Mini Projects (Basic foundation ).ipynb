# 2. Cinema Booking System

## A cinema hall has the following ticket pricing:

Seat Type:
 
  Type 1 → Normal   → ₹120 per ticket
  
  Type 2 → Premium  → ₹200 per ticket
  
  Type 3 → Recliner → ₹350 per ticket

Day discount:
  
  Monday to Thursday → 20% discount on ticket price
  
  Friday             → No discount
  
  Saturday & Sunday  → 15% extra charge added

Bulk booking:
  
  5 or more tickets  → additional 5% off final price

Take seat type, day, and number of tickets as input.

Print price per ticket, discount or extra charge,

bulk discount if any, and total amount.

```python
seat_type = int(input("Enter the seat type (1/2/3): "))
if seat_type == 1:
    seat_name = "Normal"
    price = 120
elif seat_type == 2:
    seat_name = "Premium"
    price = 200
elif seat_type == 3:
    seat_name = "Reclyner"
    price = 350
else:
    print("Invalid Seat Name!")
    exit()
    
    
day = input("Enter Day : ")
if day in ["Monday","Tuesday","Wednesday","Thursday"]:
    new_price = price - (price * 20/100)
elif day == "Friday":
    new_price = price
elif day in ["Saturday" ,"Sunday"]:
    new_price = price + (price * 15/100)
else:
    print("Invalid Input")
    exit()

tickets = int(input("Enter the Number of tickets: "))
subtotal = new_price * tickets
discount = price - new_price
if tickets >= 5:
    bulk_discount = subtotal * 5/100
    total = subtotal - bulk_discount
else:
    bulk_discount = 0
    total = subtotal

print("=====================================================")
print("              CINEMA BOOKING SYSTEM                  ")
print("=====================================================")
print(f"Seat Type : {seat_name}")
print(f"Base Price : {price}")
print(f"Day : {day}")
print(f"Weekday Discount : {discount}")
print(f"Price After Discount : {new_price}")
print(f"Number of Tickets : {tickets}")
print(f"Sub Total : ₹{subtotal}  ")
print(f"Bulk Discount : {bulk_discount}")
print("=====================================================")
print(f"Total : ₹{total}")
print("=====================================================")
```

# 3. Employee Salary Calculator

## A company calculates employee salary with the following rules:

Basic Salary components:
 
  HRA  (House Rent Allowance) → 20% of basic
  
  DA   (Dearness Allowance)   → 15% of basic
  
  TA   (Travel Allowance)     → ₹1500 fixed always

Deductions:
  
  PF   (Provident Fund)       → 12% of basic
  
  Tax  based on Gross Salary:
  
       Gross < ₹20,000        → No tax
       
       ₹20,000 to ₹40,000     → 5% tax on gross
       
       ₹40,001 to ₹80,000     → 10% tax on gross
       
       Above ₹80,000          → 20% tax on gross

Gross Salary = Basic + HRA + DA + TA

Net Salary   = Gross - PF - Tax

Take employee name and basic salary as input.

Print full salary slip.

```python
name = input("Enter Employee Name : ")
basic = float (input("Enter Basic Salary : "))

hra = basic * 20/100
da = basic * 15/100
ta = 1500.00
pf = basic * 12/100
gross = basic + hra + da + ta 

if gross <= 20000:
    tax = 0
elif gross <= 40000:
    tax = gross * 5/100
elif gross <=80000:
    tax = gross * 10/100
elif gross >80000:
    tax = gross * 20/100
else:
    print("Invalid Data !")
    exit()
total = pf + tax
net = gross - pf - tax

print("======================================================")
print("            EMPLOYEE SALARY CALCULATOR                ")
print("======================================================")
print(f"Employee Name : {name}")
print(f"Basic Salary  :  {basic}")
print("------------------------------------------------------")
print("                      EARNINGS                        ")
print("------------------------------------------------------")
print(f"Basic Salary     : {basic}")
print(f"HRA (20 %)       : {hra}")
print(f"DA  (15 %)       : {da}")
print(f"TA (fixed)       : {ta}")
print("------------------------------------------------------")
print(f"GROSS SALARY     :  {gross} ")
print("------------------------------------------------------")
print("------------------------------------------------------")
print("                     DEDUCTION                        ")
print("------------------------------------------------------")
print(f"PF (12% )        : {pf} ")
print(f"TAX (10% )       : {tax} ")
print("------------------------------------------------------")
print(f"TOTAL DEDUCTION  :  {total}                    ")
print("------------------------------------------------------")
print(f"NET SALARY       :  {net}")
print("======================================================")
```

# 4.Road Trip Fuel Cost Calculator

## A person is planning a road trip and wants to calculate the total fuel cost.

Fuel price varies by fuel type:

  Petrol  → ₹102.50 per litre
  
  Diesel  → ₹89.00  per litre
  
  CNG     → ₹75.00  per kg

Mileage of vehicle varies by type:
  
  Car     → 15 km per litre/kg
  
  Bike    → 45 km per litre/kg
  
  Truck   → 8  km per litre/kg

Extra toll charges based on distance:
  
  Up to 100 km    → ₹0 toll
  
  101 to 300 km   → ₹150 toll
  
  301 to 600 km   → ₹350 toll
  
  Above 600 km    → ₹600 toll

Take fuel type, vehicle type, and distance as input.

Print fuel needed, fuel cost, toll charge,

and total trip cost.

```python
fuel_type = input("Enter Fuel Type (petrol/diesel/cng) : ")
vehicle_type = input("Enter vehicle type (car/bike/truck) : ")
distance = float(input("Enter distance (km) : "))

if fuel_type == "Petrol":
    fuel_price = 102.50 
elif fuel_type == "Diesel":
    fuel_price = 89.00
elif fuel_type == "Cng":
    fuel_price = 75.00
else:
    print("Invalid Input!")
    exit()

if vehicle_type == "Car":
    mileage = 15 
elif vehicle_type == "Bike":
    mileage = 45
elif vehicle_type == "Truck":
    mileage = 8
else:
    print("Invalid Input!")
    exit()
    
fuel_needed = distance / mileage
fuel_cost = fuel_needed * fuel_price

if distance <= 100:
    toll_charge = 0
elif distance <= 300:
    toll_charge = 150
elif distance <= 600:
    toll_charge = 350
elif distance > 600:
    toll_charge = 600
else:
    print("Invalid Input!")
    exit()

trip_cost = fuel_cost + toll_charge

print("==============================================================")
print("                  ROAD TRIP COST CALCULATOR                   ")
print("==============================================================")
print(f" Fuel Type                : {fuel_type}")
print(f" Vehicle Type             : {vehicle_type}")
print(f" Distance                 : {distance}")
print(f" Mileage                  : {mileage}")  
print("--------------------------------------------------------------")
print(f" Fuel Required            : {fuel_needed}")
print(f" Fuel Price               : {fuel_price}")
print(f" Fuel Cost                : {fuel_cost}")
print(f" Toll Charge              : {toll_charge}")
print("--------------------------------------------------------------")
print(f" Total Trip Cost          : {trip_cost}")
print("==============================================================")
```

# 5 . Hotel Room Booking System

## A hotel charges based on room type and number of nights with the following pricing:

Room Types:
  
  Type 1 → Standard  → ₹1500 per night
  
  Type 2 → Deluxe    → ₹2500 per night
  
   Type 3 → Suite     → ₹5000 per night

Stay discount based on nights:
 
  1 to 2 nights  → No discount
  
  3 to 6 nights  → 10% discount on room total
  
  7 or more      → 20% discount on room total

Meal plan (added to total after discount):
  
  Option 1 → No meals      → ₹0
  
  Option 2 → Breakfast     → ₹250 per person per night
  
  Option 3 → All inclusive → ₹600 per person per night

GST on final amount → 12%

Take room type, nights, number of guests and meal plan as input.

Print full booking summary.

```python
room_type = input("Enter room type  (Standard/Deluxe/Suite) : ")
number_of_nights = int(input("Enter number of nights : "))
number_of_guests = int(input("Enter number of guests : "))
meal_plan = input("Enter meal plan  (No meals/Breakfast/All Inclusive) : ")

price_per_night = 0
if room_type == "Standard":
    price_per_night = 1500
elif room_type == "Deluxe":
    price_per_night = 2500
elif room_type == "Suite":
    price_per_night = 5000
else:
    print("Invalid Input !")
    exit()

room_total = price_per_night * number_of_nights

if number_of_nights <= 0:
    print("Invalid number of  nights ")
    exit()
elif number_of_nights <= 2:
    discount = 0
elif number_of_nights <= 6:
    discount =  room_total * 10 / 100
elif number_of_nights >= 7:
    discount = room_total * 20 / 100


after_discount = room_total - discount 

meal_cost = 0
if meal_plan == "No meals":
    meal_cost = 0
elif meal_plan == "Breakfast":
    meal_cost = number_of_nights * 250
elif meal_plan == "All Inclusive":
    meal_cost = number_of_nights * 600
else:
    print("Invalid Input !")
    exit()

sub_total = after_discount + meal_cost
 
gst = sub_total * 12 / 100
total = sub_total + gst 
 

print("=========================================================")
print("               Hotel Room Booking System                 ")
print("=========================================================")
print(f"Room Type           : {room_type}")
print(f"Price Per Night     : {price_per_night}")
print(f"Number of NIghts    : {number_of_nights}")
print(f"Number of Guests    : {number_of_guests}")
print("---------------------------------------------------------")
print(f"Room Total          : {room_total}")
print(f"Stay Discount       : {discount}")
print(f"After Discount      : {after_discount}")
print("---------------------------------------------------------")
print(f"Meal Plan           : {meal_plan}")
print(f"Meal Cost           : {meal_cost}")
print("---------------------------------------------------------")
print(f"Sub Total           : {sub_total}")
print(f"GST                 : {gst}")
print("---------------------------------------------------------")
print(f"GRAND TOTAL         : {total}")
print("=========================================================")
print(" THNAK YOU ! ENJOY YOUR STAY 🏠:")
print("=========================================================")
```
