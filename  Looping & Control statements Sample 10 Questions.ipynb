# Sum of Even numbers 

Find sum of EVEN numbers from 5 inputs”

You only need:

Take number

Check even/odd

Add to total

```python
sum=0
for i in range(5):
    num=int(input("Enter the Number: "))
    if num%2 == 0:
        sum+=num
print("Sum of the Even Number:",sum)
```

# 8.Find the AVERAGE of numbers that are positive out of 5 inputs

```python
count=0
total=0

n=int(input("Enter the Number :"))

for i in range(n):
    num=int(input("Enter the value: "))
    
    if num>0:
        total += num
        count += 1
        
if count>0:
    average =total/count
    print("Average: ",average)
else:
    print("No positive numbers")
```
